```python
from konlpy.tag import Hannanum
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
url = 'https://www1.president.go.kr/articles/7940'
html = requests.get(url)
soup = BeautifulSoup(html.text, 'html.parser')
text = soup.select_one('.cs_body > .cs_view.text > .text')
text = text.get_text(' ', strip=True)
text
```

```python
text = re.sub('[0-9]+', '', text)
text = re.sub('[A-Za-z]+', '', text)
text = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ·!』\\‘’|\(\)\[\]\<\>`\'…》]', '', text)
```

```python
hannanum = Hannanum()
text_list = hannanum.nouns(text)
text_list
```

```python
word_list = pd.Series(text_list)
result = word_list.value_counts().head(20)
result
```

```python
dfsub = df[df['순위']>=10][df['날짜']>'2020-04-01'][df['법인구분'].str.find('시장') > 0]
text = str.join(',', [x['업종명']+' '+x['회사이름']+'\n' for x in dfsub.iloc])
print(len(text))
```

```python
text = re.sub('[0-9]+', '', text)
text = re.sub('[A-Za-z]+', '', text)
text = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ·!』\\‘’|\(\)\[\]\<\>`\'…》]', '', text)

texttemp = text
text_list = list()
hannanum = Hannanum()
while len(texttemp) > 10000:
    cut = texttemp[:10000].rfind('\n')
    text_list += hannanum.nouns(texttemp[:cut])
    texttemp = texttemp[cut:]
cut = texttemp[:10000].rfind('\n')
text_list += hannanum.nouns(texttemp[:cut])
texttemp = texttemp[cut:]
```

```python
print(len(text_list))
word_list = pd.Series(text_list)
result = word_list.value_counts()
result

for i, k in enumerate(result.keys()):
    print(i, k, result[i])
```

```python

import os
import shutil
def move(root_src_dir, root_dst_dir):
    try:
        for src_dir, dirs, files in os.walk(root_src_dir):
            dst_dir = src_dir.replace(root_src_dir, root_dst_dir, 1)
            if not os.path.exists(dst_dir):
                os.makedirs(dst_dir)
            for file_ in files:
                src_file = os.path.join(src_dir, file_)
                dst_file = os.path.join(dst_dir, file_)
                if os.path.exists(dst_file):
                    # in case of the src and dst are the same file
                    if os.path.samefile(src_file, dst_file):
                        continue
                    os.remove(dst_file)
    #             print(src_file, dst_dir)
                shutil.move(src_file, dst_dir)
            if not os.listdir(src_dir):
    #             print(src_dir)
                os.removedirs(src_dir)
        return True
    except Exception as e:
        print(e)
        return False
```
